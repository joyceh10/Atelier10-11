### Exercice 4.4 
Creer une fonction  appelée 'age_group' qui catégorise l'âge en groupes :
 - Moins de 20 ans : "0-19"
 - Entre 20 et 40 ans : "20-39"
 - Entre 40 et 60 ans : "40-59"
 - Plus de 60 ans : "60+"
 Créez une colonne "age_group" de catégorie des âges. Appliquer la fonction age_group à la colonne 'age' pour créer la nouvelle colonne 'age_group' avec df["nom de la colonne a modifier"].apply(age_group)

```python
def age_group(age):
    if age < 20:
        return "0-19"
    elif 19 < age < 40 :
        return "20-39"
    elif 39 < age < 50:
        return "40-59"
    else:
        return "60+"
df ['age_group'] = df['age'].apply(age_group)
```
