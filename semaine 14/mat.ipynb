## Exemple 2: deux courbe

```python
import matplotlib.pyplot as plt
# Données
x = [1, 2, 3, 4, 5]
y1 = [10, 20, 25, 30, 40]
y2 = [5, 15, 20, 25, 35]
# Création de la figure avec taille personnalisée (en cm converti en pouces)
fig, ax = plt.subplots(figsize=(20/2.54, 
12/2.54)) # ~20 cm x 12 cm
# Courbe 1
ax.plot(x, y1, color='blue', linestyle='-', 
linewidth=2, marker='o', label='Série A')
# Courbe 2
ax.plot(x, y2, color='green', linestyle='--', 
linewidth=2, marker='s', label='Série B')
# Titre et axes
ax.set_title("Comparaison de deux séries", 
fontsize=14, fontweight='bold')
ax.set_xlabel("Temps (s)", fontsize=12)
ax.set_ylabel("Valeur mesurée", fontsize=12)

# Limites des axes
ax.set_xlim(0, 6)
ax.set_ylim(0, 50)
# Grille personnalisée
ax.grid(True, linestyle='--', color='gray', 
alpha=0.7)
# Légende
ax.legend(loc='upper left', fontsize=10)
# Ajustement automatique des marges
plt.tight_layout()
# Sauvegarde du graphique
plt.savefig("graphique_complet.png", dpi=300)
# Affichage
plt.show()
```

```python
import matplotlib.pyplot as plt
# notre fonction est: f(x) = x² - 2x + 1
# Définir l'intervalle des x
x = list(range(-10, 11))
# Calculer f(x) = x² - 2x + 1 pour chaque x
y = [xi**2 - 2*xi + 1 for xi in x]
# Créer la figure et l'axe
fig, ax = plt.subplots()
# Tracer la fonction
ax.plot(x, y, label="f(x) = x² + 2x + 1", color="blue")
# Ajouter un titre et des labels
ax.set_title("Courbe de la fonction f(x) = x² + 2x + 1")
ax.set_xlabel("x")
ax.set_ylabel("f(x)")
# Ajouter une grille
ax.grid(True)
# Ajouter une légende
ax.legend()
# Afficher le graphique
plt.show()
```

```python
import matplotlib.pyplot as plt
# Données fictives sur la répartition des dépenses mensuelles
categories = ['Logement', 'Alimentation', 'Transports', 'Divertissement', 'Santé']
depenses = [40, 30, 15, 10, 5] # Pourcentages de dépenses
# Tracer le camembert sans utiliser 'explode'
plt.pie(depenses,
labels=categories,
autopct='%1.1f%%',
startangle=90,
colors=['lightblue', 'lightgreen', 'lightcoral', 'gold', 'lightgray'],
wedgeprops={'edgecolor': 'black'},
labeldistance=1.2, # Ajuster la distance des étiquettes
pctdistance=0.85) # Ajuster la distance entre le pourcentage et le centre du camembert
# Titre
plt.title("Répartition des dépenses mensuelles d'un ménage", pad=20)
# Affichage
plt.axis('equal') # Cercle parfait
plt.show()
```
